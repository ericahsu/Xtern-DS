# Correlation of Rating

Visualize the correlation between rating of a restaurant and the number of votes or reviews. See what connections are implied from the graphs.

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd


df = pd.read_csv('2020-XTern-DS.csv')
```

Clean the rating and vote columns so any rows without data are not included.

```python
df['Rating'] = df['Rating'].astype(str) 

df = df[~df['Rating'].str.contains('-')]
df = df[~df['Rating'].str.contains('NEW')]
df = df[~df['Rating'].str.contains('Opening Soon')]
df['Rating'] = df['Rating'].astype(float) 

df = df[~df['Votes'].str.contains('-')] 
df['Votes'] = df['Votes'].astype(float)

df = df[~df['Reviews'].str.contains('-')]
df['Reviews'] = df['Reviews'].astype(float)
```

Create the graph with the Rating on the x-axis and Votes on the y-axis.

```python
plt.figure(figsize=(20,10))

plt.scatter(df['Rating'], df['Votes'])
plt.xlabel("Rating")
plt.ylabel("Votes");
```

The overall trend seems to be positive. As the rating decreases, the number of votes decreases. When the rating increases, the votes also increase, but there also appear to be a lot of highly rated restaurants that also have a low number of votes. The restaurants with the highest rating actually have less votes than some of the restaurants in the 4.0 area. It's possible that the highest restaurants have a high rating because there isn't enough of a sample from the votes yet.

Create the graph with the Rating on the x-axis and Reviews on the y-axis.

```python
plt.figure(figsize=(20, 10))

plt.scatter(df['Rating'], df['Reviews'])
plt.xlabel("Rating")
plt.ylabel("Reviews");
```

Looking at the correlation between the rating and number of reviews, a similar pattern can be seen as in the graph above. Most of the restaurants with a rating of less than 3.5 also have a low number of reviews, though there are some outliers. The restaurants with a rating between 3.5 to 4.25 have reviews ranging from the minimum to the maximum amount. Restaurants with ratings above 4.25 have a similar range in reviews as the 3.5 to 4.25 ratings, but it's more sparse. This could be due to there being fewer restaurants with ratings in that range, rather than actually being a correlation between reviews and ratings.
